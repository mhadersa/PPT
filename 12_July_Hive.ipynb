```python
from pyhive import hive

# Establish a connection to Hive
connection = hive.Connection(host='localhost', port=10000, username='mansa12')

# Create the Employees table
create_table_query = """
CREATE TABLE Employees (
    id INT,
    name STRING,
    salary FLOAT
)
"""

# Execute the create table query
with connection.cursor() as cursor:
    cursor.execute(create_table_query)

# Close the connection
connection.close()
```

```python
from pyhive import hive

# Establish a connection to Hive
connection = hive.Connection(host='localhost', port=10000, username='mansa12')

# Define the HiveQL query
query = """
SELECT *
FROM Customers
WHERE age > 30
"""

# Execute the query
with connection.cursor() as cursor:
    cursor.execute(query)
    results = cursor.fetchall()

# Print the results
for row in results:
    print(row)

# Close the connection
connection.close()
```

```python
from pyhive import hive

# Establish a connection to Hive
connection = hive.Connection(host='localhost', port=10000, username='mansa12')

# Define the HiveQL query
query = """
SELECT AVG(salary) AS avg_salary
FROM Employees
"""

# Execute the query
with connection.cursor() as cursor:
    cursor.execute(query)
    result = cursor.fetchone()

# Extract the average salary from the result
average_salary = result[0]

# Print the average salary
print("Average Salary:", average_salary)

# Close the connection
connection.close()
```

```python
from pyhive import hive

# Establish a connection to Hive
connection = hive.Connection(host='localhost', port=10000, username='mansa12')

# Define the HiveQL query for inner join
query = """
SELECT *
FROM Orders
JOIN Customers ON Orders.customer_id = Customers.customer_id
"""

# Execute the query
with connection.cursor() as cursor:
    cursor.execute(query)
    results = cursor.fetchall()

# Print the results
for row in results:
    print(row)

# Close the connection
connection.close()
```
